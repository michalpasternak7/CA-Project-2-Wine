```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

## Distribution Plots 

```python
# get a distribution plot to show the distribution of a univariate set
sns.distplot(tips['total_bill'])
```

```python
# get a distribution plot to show the distribution of a univariate set
sns.distplot(tips['total_bill'], kde = False)
```

```python
# get a distribution plot to show the distribution of a univariate set
sns.distplot(tips['total_bill'], kde = False, bins= 30)
```

```python
# jointplot: essentially combine two different distribution plots. two variables.
sns.jointplot(x= 'total_bill' , y= 'tip', data=tips )
```

```python
# jointplot: essentially combine two different distribution plots. two variables.
sns.jointplot(x= 'total_bill' , y= 'tip', data=tips, kind = 'hex' )
```

```python
# jointplot: essentially combine two different distribution plots. two variables.
sns.jointplot(x= 'total_bill' , y= 'tip', data=tips, kind = 'reg' )
```

```python
# jointplot: essentially combine two different distribution plots. two variables.
sns.jointplot(x= 'total_bill' , y= 'tip', data=tips, kind = 'kde' )
```

```python
# Pairplot to plot pairwise relationships across an entire data frame at least for the numerical columns.
# its joint plots for every single possible combination of the numerical columns in this data frame.
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue = 'sex')
```

```python
sns.pairplot(tips, hue = 'sex', palette= 'coolwarm')
```

```python
# rugplot to draw a dash mark for every single point along the distribution line.
sns.rugplot(tips['total_bill'])
```

```python
#  build the this the KDE plot without any distribution of the bars.
sns.kdeplot(tips['total_bill'])
```

## Categorical Plots

```python
# for categorical plots were mainly going to be concerned about seeing the distributions of a categorical column 
```

```python
# barplot will create a bar plots showing  the average or mean of the total bill.
sns.barplot( data= tips, x= 'sex', y= 'total_bill') 
```

```python
import numpy as np
```

```python
# use an estimator to see what is the standard deviation of the total bill column per category
sns.barplot(x= 'sex', y= 'total_bill', data= tips, estimator= np.std)
```

```python
# count plots is essentially the same as a bar plot except the estimator 
# is explicitly counting the number of occurrences. so, we only set the x value. 
sns.countplot(x= 'sex', data = tips)
```

```python
# box plot shows the distribution of quantitative data in a way that facilitates comparisons between the variables.
sns.boxplot(x= 'day', y= 'total_bill', data = tips)
```

```python
# get a bit fancy 
sns.boxplot(x= 'day', y= 'total_bill', data = tips, hue= 'smoker')
```

```python
# Now the violin plot unlike the box plot allows us to actually plot all the components that correspond
# to actual data points and it's essentially showing the kernel density estimation of the underlying distribution
sns.violinplot(x= 'day', y= 'total_bill', data = tips)
```

```python
sns.violinplot(x= 'day', y= 'total_bill', data = tips, hue = 'sex')
```

```python
# advanced visualisation 
sns.violinplot(x= 'day', y= 'total_bill', data = tips, hue = 'sex', split = True)
```

```python
#  the strip plot is going to draw a scatterplot where one variable is categorical.
sns.stripplot(x= 'day', y= 'total_bill', data = tips, jitter= False)
```

```python
# So the normal stript plot if it had a bunch of points stacked on top of each other its hard to tell.
# so to see the density a little better 

sns.stripplot(x= 'day', y= 'total_bill', data = tips)
```

```python
sns.stripplot(x= 'day', y= 'total_bill', data = tips, hue= 'sex')
```

```python
sns.stripplot(x= 'day', y= 'total_bill', data = tips, hue= 'sex', split = True)
```

```python
# swarm plot is essentially a combination of a strip plot and a violent plot. 
# It shows the same distribution except it's actually showing all the points themselves.
# in general probably stay away from just showing swarm plots since most people probably haven't encountered them before.

sns.swarmplot(x= 'day', y= 'total_bill', data = tips)
```

```python
# not obvious? okay here we go! :) 
#This is probably not useful for presentation but maybe just for exploratory data analysis 
sns.violinplot(x= 'day', y= 'total_bill', data = tips)
sns.swarmplot(x= 'day', y= 'total_bill', data = tips, color= 'black')
```

```python
sns.factorplot(x= 'day' , y= 'total_bill', data = tips, kind= 'bar' )
```
